```python
# key to tropename
tropes = pd.read_csv('dataset/my_tropes.csv')
def tid_to_tname(tids):
    tnames = []
    if not tids:
        return []
    for tid in tids:
        t = tropes[tropes['TropeID'] == tid].iloc[0]['Trope']
        tnames.append(t)
    return tnames
```

```python
stories
```

```python
# sids = tropes_sim.columns[4:]
def get_tropes_for_story(sid):
    title = stories[stories['sid']==sid].iloc[0]['Title']
    print(sid, title)
    filtered_tropes = tropes_sim_story[tropes_sim_story[sid]>0.65][['trope_id', 'example_type', sid]]
    # sort by similarity values
    filtered_tropes = filtered_tropes.sort_values(sid, ascending=False)
    gp = filtered_tropes.groupby('trope_id', sort=False)
    keys = [key for key, _ in gp]
    # print(keys)
    trope_names = tid_to_tname(keys)
    # get top 10 "unique" tropes since tropes will be repeated and may also have different values
    print(trope_names[:10])
    print("#############################################")
    return keys[:10]
```

```python
stories['tropes'] = stories['sid'].progress_apply(get_tropes_for_story)
stories
```
